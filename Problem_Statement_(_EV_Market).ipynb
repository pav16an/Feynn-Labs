```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
```

```python
cols = ['Model', 'Manufacturer', 'Vehicle Type', 'Battery Capacity (kWh)',
       'Range per Charge (km)', 'Charging Time', 'Price', 'Power (HP or kW)',
       'Top Speed (km/h)', 'Year of Manufacture']
df[cols] = df[cols].apply(LabelEncoder().fit_transform)
df.head()
```

```python
features = list(df.columns)
d = df[features]
```

```python
features
```

```python
from sklearn.cluster import KMeans
clustering_kmeans = KMeans(n_clusters=3)
d['clusters'] = clustering_kmeans.fit_predict(d)
```

```python
d
```

```python
k_range = range(1,10)
#sum of squared error
sse = []
for k in k_range:
  km = KMeans(n_clusters=k)
  km.fit(d)
  sse.append(km.inertia_)
```

```python
sse
```

```python
plt.title('The Elbow Point Graph')
plt.xlabel('K')
plt.ylabel('sse')
plt.plot(k_range,sse)
```

```python
clustering_kmeans = KMeans(n_clusters=3)
d['clusters'] = clustering_kmeans.fit_predict(d)
d
```

```python
def plotseg(clus,Y, labels):
  plt.scatter(clus[Y==0,0], clus[Y==0,1], s=10, c='blue', label='Cluster 1')
  plt.scatter(clus[Y==1,0], clus[Y==1,1], s=10, c='red', label='Cluster 2')
  plt.scatter(clus[Y==2,0], clus[Y==2,1], s=10, c='black', label='Cluster 3')

  plt.title('Customer_segments')
  plt.xlabel(labels[0])
  plt.ylabel(labels[1])
  plt.show()
```

```python
clus = d.loc[:,["Manufacturer","Price"]].values
kmeans = KMeans(n_clusters=3, init='k-means++')
Y = kmeans.fit_predict(clus)
plotseg(clus, Y, ["Manufacturer","Price"])
```

```python
clus = d.loc[:,["Battery Capacity (kWh)","Price"]].values
kmeans = KMeans(n_clusters=3, init='k-means++')
Y = kmeans.fit_predict(clus)
plotseg(clus, Y, ["Battery Capacity (kWh)","Price"])
```

```python
clus = d.loc[:,["Range per Charge (km)","Price"]].values
kmeans = KMeans(n_clusters=3, init='k-means++')
Y = kmeans.fit_predict(clus)
plotseg(clus, Y, ["Range per Charge (km)","Price"])
```

```python
clus = d.loc[:,["Charging Time","Price"]].values
kmeans = KMeans(n_clusters=3, init='k-means++')
Y = kmeans.fit_predict(clus)
plotseg(clus, Y, ["Charging Time","Price"])
```
